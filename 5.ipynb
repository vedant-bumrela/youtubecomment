```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris


df = load_iris()
iris_data = pd.DataFrame(df.data, columns=df.feature_names)
iris_data['target'] = df.target
iris_data['species'] = iris_data['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})


print("Missing values in each column:")
print(iris_data.isnull().sum())


print("\nSummary statistics of the data:")
print(iris_data.describe())


sns.set(style="whitegrid")
iris_data.drop('species', axis=1).hist(bins=15, figsize=(10, 8))
plt.suptitle('Distribution of Features', fontsize=16)
plt.show()

correlation_matrix = iris_data.drop(['species', 'target'], axis=1).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap", fontsize=16)
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(data=iris_data.drop(['species', 'target'], axis=1))
plt.title("Box Plot for Outliers", fontsize=16)
plt.show()


sns.pairplot(iris_data, hue="species", palette="Set1")
plt.suptitle("Pairplot of Features", fontsize=16)
plt.show()


plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=iris_data, palette="Set1")
plt.title("Species Distribution", fontsize=16)
plt.show()


sns.pairplot(iris_data, hue="target", palette="viridis", markers=["o", "s", "D"])
plt.suptitle("Pairplot with Target Variable", fontsize=16)
plt.show()
```
